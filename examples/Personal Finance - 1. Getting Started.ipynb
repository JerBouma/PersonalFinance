```python
from financeportfolio.cashflows import cashflow_controller
```

```python
cf = cashflow_controller.Cashflow('examples/configurations/cashflow_example.yaml')
```

```python
cf.perform_analysis()
```

You can view the transactions by calling the related DataFrame.

```python
print("Weekly Overview")
display(cf._weekly_overview)

print("Monthly Overview")
display(cf._monthly_overview)

print("Quarterly Overview")
display(cf._quarterly_overview)

print("Yearly Overview")
display(cf._yearly_overview)
```

Next to that, you can view all related period datasets that are categorized per week, month, quarter and year.

```python
print("Weekly Cash Flows")
display(cf._weekly_cash_flow_dataset)

print("Monthly Cash Flows")
display(cf._monthly_cash_flow_dataset)

print("Quarterly Cash Flows")
display(cf._quarterly_cash_flow_dataset)

print("Yearly Cash Flows")
display(cf._yearly_cash_flow_dataset)
```
